```python
# ==========================================
# 1️⃣ Import Libraries
# ==========================================
import pandas as pd
import numpy as np
import pickle
import os
import json
from groq import Groq
from dotenv import load_dotenv

# ==========================================
# 2️⃣ Load ENV
# ==========================================
load_dotenv()
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

# ==========================================
# 3️⃣ Load Model + Columns
# ==========================================
model = pickle.load(open("artifacts/xgb_churn_model.pkl", "rb"))
model_columns = pickle.load(open("artifacts/model_columns.pkl", "rb"))

# ==========================================
# 4️⃣ Load SHAP Values
# ==========================================
with open("artifacts/shap_values.pkl", "rb") as f:
    shap_values = pickle.load(f)

print("✅ SHAP values loaded!")

# ==========================================
# 5️⃣ Load Test Data
# ==========================================
test_df = pd.read_csv("datasets/test_data.csv")

X_test_raw = test_df.drop("Customer Status", axis=1)
y_test = test_df["Customer Status"]

# ==========================================
# 6️⃣ Encode + Align Columns
# ==========================================
X_test = pd.get_dummies(X_test_raw)

for col in model_columns:
    if col not in X_test:
        X_test[col] = 0

X_test = X_test[model_columns]

# ==========================================
# 7️⃣ Get Top SHAP Features
# ==========================================
def get_top_features(customer_index, class_index=1, top_n=5):

    shap_vals = shap_values.values[customer_index, :, class_index]

    df = pd.DataFrame({
        "feature": model_columns,
        "value": X_test.iloc[customer_index].values,
        "shap_value": shap_vals
    })

    df["importance"] = np.abs(df["shap_value"])
    df = df.sort_values("importance", ascending=False)

    return df.head(top_n)

# ==========================================
# 8️⃣ Build Prompt (STRICT JSON)
# ==========================================
def build_prompt(customer_data, top_features, churn_prob):

    features_text = ""

    for _, row in top_features.iterrows():
        direction = "increases churn" if row['shap_value'] > 0 else "decreases churn"

        features_text += f"{row['feature']}={row['value']} ({direction}, impact={row['shap_value']:.3f})\n"

    prompt = f"""
You are a telecom churn expert.

Churn Probability: {round(churn_prob*100,2)}%

Customer:
{customer_data.to_dict()}

Top factors:
{features_text}

RULES:
- If feature increases churn → action MUST be "decrease"
- If feature decreases churn → action MUST be "increase"
- Keep change small and realistic

Return ONLY JSON:
[
  {{
    "strategy": "short title",
    "feature": "EXACT feature name",
    "action": "increase or decrease",
    "change": number,
    "description": "short explanation"
  }}
]
"""

    return prompt

# ==========================================
# 9️⃣ Generate Strategies
# ==========================================
def generate_strategies(customer_index):

    top_features = get_top_features(customer_index)

    churn_prob = model.predict_proba(X_test.iloc[[customer_index]])[0][1]

    prompt = build_prompt(X_test_raw.iloc[customer_index], top_features, churn_prob)

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": "You are a telecom retention strategist."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.3
    )

    raw_output = response.choices[0].message.content

    print("\n🔍 RAW LLM OUTPUT:\n", raw_output)

    try:
        strategies = json.loads(raw_output)
        print("\n✅ Parsed Strategies:\n", strategies)
    except:
        print("\n❌ JSON PARSE FAILED")
        strategies = []

    return strategies, churn_prob, top_features

# ==========================================
# 🔟 TEST RUN
# ==========================================
customer_index = 0

strategies, churn_prob, top_features = generate_strategies(customer_index)

print("\n🔥 ===== FINAL OUTPUT ===== 🔥\n")
print("Churn Probability:", round(churn_prob * 100, 2), "%")
print("\nTop Features:\n", top_features)
print("\nStrategies:\n", strategies)
```
