```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df["Offer"].fillna("None", inplace=True)
```

```python
df["Avg Monthly Long Distance Charges"] = df["Avg Monthly Long Distance Charges"].fillna(0)
```

```python
df["Internet Type"].fillna("No Internet", inplace=True)
```

```python
df["Avg Monthly GB Download"] = df["Avg Monthly GB Download"].fillna(0)
```

```python
print(df.columns.tolist())
```

```python
df["Online Security"].replace(" ", pd.NA, inplace=True)
df["Online Security"].fillna("No Internet", inplace=True)

df["Online Backup"].fillna("No Internet", inplace=True)
df["Device Protection Plan"].fillna("No Internet", inplace=True)
df["Premium Tech Support"].fillna("No Internet", inplace=True)
df["Streaming TV"].fillna("No Internet", inplace=True)
df["Streaming Movies"].fillna("No Internet", inplace=True)
```

```python
df["Churn Category"] = df["Churn Category"].fillna("No Churn")
df["Churn Reason"] = df["Churn Reason"].fillna("No Churn")
```

```python
df["Offer"] = df["Offer"].fillna("None")
df["Internet Type"] = df["Internet Type"].fillna("No Internet")
df["Multiple Lines"] = df["Multiple Lines"].fillna("No Phone")
df["Avg Monthly Long Distance Charges"] = df["Avg Monthly Long Distance Charges"].fillna(0)
df["Avg Monthly GB Download"] = df["Avg Monthly GB Download"].fillna(0)

df["Churn Category"] = df["Churn Category"].fillna("No Churn")
df["Churn Reason"] = df["Churn Reason"].fillna("No Churn")
service_cols = [
    "Online Security",
    "Online Backup",
    "Device Protection Plan",
    "Premium Tech Support",
    "Streaming TV",
    "Streaming Movies",
    "Streaming Music"
]

for col in service_cols:
    df[col] = df[col].replace(" ", pd.NA)
    df[col] = df[col].fillna("No Internet")

```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
y = df["Customer Status"]

# Feature set
X = df.drop("Customer Status", axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le_target = LabelEncoder()

y_encoded = le_target.fit_transform(df["Customer Status"])
```

```python
X.select_dtypes(include="object").columns
print(X.columns.tolist())
```

```python
X = X.drop("Customer ID", axis=1)
```

```python
X = X.drop(["Churn Reason", "Churn Category"], axis=1)
```

```python
categorical_cols = X.select_dtypes(include="object").columns
numeric_cols = X.select_dtypes(exclude="object").columns
categorical_cols
```

```python
categorical_cols
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import pickle
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_cols)
    ],
    remainder="passthrough"
)
X_encoded = preprocessor.fit_transform(X)
X_encoded


# Save preprocessor
pickle.dump(preprocessor, open("artifacts/preprocessor.pkl", "wb"))
```

Convert output to DataFrame with column names

```python
# Get encoded column names
encoded_cols = preprocessor.named_transformers_["cat"].get_feature_names_out(categorical_cols)

# Combine with numeric column names
all_cols = list(encoded_cols) + list(numeric_cols)

# Create final dataframe
X_encoded_df = pd.DataFrame(X_encoded, columns=all_cols)

print(X_encoded_df.head())
print(X_encoded_df.columns.tolist())
```

```python
X_encoded_df.describe
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded_df,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y   # VERY IMPORTANT for imbalance
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
print(X_train.columns.tolist())
```

```python
train_df = X_train.copy()
train_df["Customer Status"] = y_train

train_df.to_csv("datasets/train_data.csv",
                index=False,
                chunksize=1000)   # write in chunks
test_df = X_test.copy()
test_df["Customer Status"] = y_test

test_df.to_csv("datasets/test_data.csv",
               index=False,
               chunksize=1000)
```
